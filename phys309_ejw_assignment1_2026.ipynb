### Questions


Recall that the speed of light in vacuum is 
\begin{equation}
c = \frac{1}{\sqrt{\epsilon_0 \mu_0}}, \; \; \; \; \; (1)
\end{equation}
where $\epsilon_0 = 8.85 \times 10^{-12}$ F/m (Farads per meter)$^{*}$ is the electric permittivity of free space and $\mu_0 = 1.26 \times 10^{-6}$ N/A$^2$ (Newtons per Ampere$^2$)$^{**}$ is the magnetic permeability of free space. The electric field created by a point charge $q$ is
\begin{equation}
E = \frac{1}{4 \pi \epsilon_0} \frac{q}{r^2}. \; \; \; \; (2)
\end{equation}
Suppose a charged particle is moving at some fraction of the speed of light in vacuum ($v_q = f c; \; 0 < f < 1$) as seen in reference frame $S$. Reference frame $S^{\prime}$ is co-moving with the point charge, so that $\frac{dx^{\prime}}{dt^{\prime}} = v^{\prime}_q = 0$ in frame $S^{\prime}$. To make the math easier, put $q$ at the origin of $S^{\prime}$.

To understand Einstein's motivation for developing the special theory of relativity, make the <strong>incorrect, erroneous, and wrong</strong> assumption that the Galilean velocity transformation holds true for all values of $f$:
\begin{equation}
\frac{dx^{\prime}}{dt^{\prime}} = \frac{dx}{dt} - v_q \; \; \; (3)
\end{equation}

<ol>
    <li>Assume the charged particle is an electron. Using Equation 2, calculate and plot the electric field as a function of distance $r^{\prime}$ from $q$ in frame $S^{\prime}$. Choose an x-axis range that makes sense to you and label both axes with correct units.</li>
</ol>

```python
# Plot here

import numpy as np
import matplotlib.pyplot as plt
from scipy import constants

# Electron charge (C). 
q = -constants.e

# Distance from the charge in frame S' (m), avoiding r' = 0 (logarthymic)
r_prime = np.logspace(-12, -8, 400)  # 1e-12 m to 1e-8 m

# Electric field magnitude from a point charge: E = (1/(4*pi*epsilon0)) * |q| / r'^2
E = (1 / (4 * np.pi * constants.epsilon_0)) * np.abs(q) / (r_prime**2)

plt.figure()
plt.loglog(r_prime, E)
plt.xlabel("Distance from Charge, r' (m)")
plt.ylabel("Electric Field Magnitude, |E(r')| (N/C)")
plt.title("Electric Field of an Electron in Frame S'")
plt.grid(True, which="both")
plt.show()
```

```python
# Plot here

import numpy as np
import matplotlib.pyplot as plt
from scipy import constants

# In frame S, the charge moves at v_q = f*c. Accodring to Galilean transformation:
# c' = c - v_q = c(1 - f)
f = np.linspace(0, 0.999, 500)        # dimensionless fraction of c
c_prime = constants.c * (1 - f)      # m/s

plt.figure()
plt.plot(f, c_prime)
plt.xlabel("Speed Fraction, f")
plt.ylabel("Incorrect Speed of Light in S', c' (m/s)")
plt.title("Incorrect Galilean Speed of Light in S' vs. Speed Fraction")
plt.grid(True)
plt.show()
```

```python
# Plot here

import numpy as np
import matplotlib.pyplot as plt
from scipy import constants

# Given: v_q = f*c in S, and Galilean: c' = c - v_q
f = np.linspace(0.0, 0.999, 500)          # Dimensionless
c_prime = constants.c * (1.0 - f)         # m/s (incorrect Galilean result)

# Eq 1 still for S': c' = 1/sqrt(epsilon0' * mu0')
# and assume mu0' = mu0
mu0_prime = constants.mu_0                # N/A^2
epsilon0_prime = 1.0 / (mu0_prime * c_prime**2)  # F/m

# Electric field in S' at r' = 1 micron due to electron
q = -constants.e                        # C
r_prime = 1e-6                            # m
E_prime = (1.0 / (4.0 * np.pi * epsilon0_prime)) * (q / r_prime**2)  # N/C

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)

# Top: epsilon' vs f
ax1.plot(f, epsilon0_prime)
ax1.set_ylabel("Permittivity ' (F/m)")
ax1.set_title("Permittivity and Electric Field vs. Galilean Speed Fraction")

# Bottom: E' at r' = 1 m vs f
ax2.plot(f, E_prime)
ax2.set_xlabel("Speed Fraction f")
ax2.set_ylabel("Electric Field E' at 1m (N/C)")

ax1.grid(True)
ax2.grid(True)

plt.tight_layout()
plt.show()
```

Recall that the magnitude of the magnetic field strength at the center of the loop in frame $S$ is $B = \frac{\mu_0 I}{2R}$, where $I$ is the current and $R$ is the radius of the loop. Suppose $I = 1$ A and $R = 1$ m. Assume $E^{\prime} = E$, $B^{\prime} = B$, and $c^{\prime}$ depends on $f$ as in problem 2. Plot $\mu_0$ as a function of $f$.</li>

<ol start=6>
<li>If we calculate $c^{\prime}$ using the Galilean transformation in Equation 3, is there any value of $f$ for which $E^{\prime} = E$ <strong>and</strong> $B^{\prime} = B$? Make a graph to illustrate your answer.</li>
</ol>

```python
# Plot here 

import numpy as np
import matplotlib.pyplot as plt
from scipy import constants

# Use a fixed r' to compare E' and E as function of f
r_prime = 1e-6          # m
q = -constants.e       # C

# f range and Galilean c', same as Prob 2
f = np.linspace(0.0, 0.999, 500)
c_prime = constants.c * (1.0 - f)

# E in S (using standard epsilon0)
E = (1.0 / (4.0 * np.pi * constants.epsilon_0)) * (q / r_prime**2)

# If Eq 1 holds and mu0' = mu0 (Prob 3) then epsilon0' changes with f:
epsilon0_prime = 1.0 / (constants.mu_0 * c_prime**2)

E_prime = (1.0 / (4.0 * np.pi * epsilon0_prime)) * (q / r_prime**2)

# Plot ratio so it's easy to see when E' = E (ratio = 1)
E_ratio = np.abs(E_prime / E)

# --- Magnetic field comparison ---
# B at center of loop in S: B = mu0 I / (2R)
I = 1.0   # A
R = 1.0   # m
B = constants.mu_0 * I / (2.0 * R)

# If we also want Eq 1 to hold while keeping epsilon0' = epsilon0 (Problem 5 idea), mu0' changes with f:
mu0_prime = 1.0 / (constants.epsilon_0 * c_prime**2)

B_prime = mu0_prime * I / (2.0 * R)
B_ratio = B_prime / B   # already positive

# Plot both ratios
plt.figure()
plt.plot(f, E_ratio, label="|E'|/|E| (Problem 3)")
plt.plot(f, B_ratio, label="B'/B (Problem 5)")
plt.axhline(1.0, linestyle="--", linewidth=1.0)
plt.xlabel("Speed Fraction, f")
plt.ylabel("Respective Ratio")
plt.title("E' = E and B' = B Ratios Under Galilean c'?")
plt.grid(True)
plt.legend()
plt.show()

#Only at f = 0 (no movement relative to observer) are the ratios equal
```
